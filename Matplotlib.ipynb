**Matplotlib** and **Pyplot** 
_Matplotlib_ is a very powerful visualization library that is useful for those want to plot data with Python.

The most used module of **Matplotlib** is **_Pyplot_**, which provides a collection of functions that let you easily plot out data.

```python
# Importing NumPy and Matplotlib.pyplot

import numpy as np
import matplotlib.pyplot as plt
```

First Example that we're going to look at using matplotlib is the **_Simple Line Graph_**. _Line graphs_ are one of the most common types of graphs.

They're generally used to plot data that is tracked over a period of time. One example of this is stock prices, which change over time.

So let's create a simple plot using _pipeline_.

```python
plt.plot([1, 7, 9, 14])
```

**Matplotlib** assumes that they are **_Y-values_** and automatically generates the X values for you.

The default X vector has the same length as the List passed into the function, but starts with zero. Here we can see the X values for the plot are zero, one, two and three.


 **_To plot X versus Y_**, You can pass two lists into the plot function. The first of which is the _X-vector_ and the second is the _Y-vector_.

```python
plt.plot([1, 2, 7, 8], [1, 9, 7, 14])
```

You can pass in a third argument to the plot function to indicate the format of the color and the line type of the plot. The default format string is **b-**, which is the solid blue line.

For example, if we want to plot the data with red circles, you would use the string R and O.

You can see the _plotlib_ documentation for a complete list of styles and format strings.

```python
plt.plot([1, 2, 7, 8], [1, 9, 7, 14], 'ro')
```

Generally, we use **NumPy** array instead of _regular python lists_ when working with **Plots** and **Matplotlib**.

```python
a = np.array([[1, 2, 3, 4],
              [5, 6, 7, 8]])

plt.plot(a[0], a[1], 'go')
```

Adding **Labels** to our **Axis**.

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(a[0], a[1], 'b--')
```

```python
t = np.linspace(0, 5, 20)

plt.plot(t, t**2, color='coral', linestyle='--')

plt.plot(t, 4*t, color='indigo', linewidth=2)
```

```python
plt.subplot(1, 3, 1)
plt.plot(t,t, 'c--')

plt.subplot(1, 3, 2)
plt.plot(t,t**2, 'b-')

plt.subplot(1, 3, 3)
plt.plot(t,t**3, 'mo')
```

We can also plot data using **Bar graphs**, which are one of the most common types of graphs.**_Bar graphs_** are used to show data that is _split into different categories_.

For example, if I want to plot out the class grades of five different students, I would use a _bar graph_ to do that.

```python
grades = {'John': 70,
          'Susan': 82,
          'Ben': 92,
          'Alex': 76,
          'David': 97}

plt.bar(range(len(grades)), list(grades.values()), color='darkturquoise')
plt.xticks(range(len(grades)), list(grades.keys()))
plt.xlabel('Students')
plt.ylabel('Grades')
```

Next Type of grapgs are **Histograms**. **_Histograms_** are a very common type of plot.

When we're looking at data like height and weight, stock prices or wait times.

```python
x = np.random.rand(1000)

plt.hist(x, color='seagreen')
```
